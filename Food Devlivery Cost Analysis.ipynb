#Food Delivery Cost Analysis using python#

**lib Import**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import csv file
df = pd.read_csv("https://raw.githubusercontent.com/gvenkatsai1919/Food-Delivery-Cost-Analysis/refs/heads/main/food%20delivery%20costs.csv")
```

```python
df.head()  #first 5 rows
```

```python
df.info() #data type
```

#Data Cleaning

```python
df["Order Date and Time"] = pd.to_datetime(df['Order Date and Time'])
df.info()
```

```python
df["Delivery Date and Time"] = pd.to_datetime(df["Delivery Date and Time"])
df.info()
```

```python
df.head()
```

```python
def extract(value):
    a = str(value).split(" ")
    return a[0]

df["Discounts and Offers"] = df["Discounts and Offers"].apply(extract)
df.head()
```

```python
def removep(value):
    if "%" in value:
        a = value.replace("%","")
        return float(a)
    else:
        return float(value)

df["Discounts and Offers"] = df["Discounts and Offers"].apply(removep)
df.head()
```

```python
df.loc[(df["Discounts and Offers"] <= 15),"Discounts and Offers" ] = (df["Discounts and Offers"]/100) * df["Order Value"]
df.head()
```

```python
df["Discounts and Offers"] = df["Discounts and Offers"].fillna(0)
df.head()
```

```python
df["Costs"] = df["Delivery Fee"] + df['Discounts and Offers'] + df["Payment Processing Fee"]  #create new columns cost(finding cost)
df.head()
```

```python
df["Profit"] = df["Commission Fee"] - df['Costs'] # create new columns profit(finding profit)
df.head()
```

```python
df["Profit"].sum() # profit sum
```

```python
cost_dist = df[["Delivery Fee", "Payment Processing Fee", "Discounts and Offers"]].sum()
cost_dist
```

```python

plt.figure(figsize = (3,3))
plt.pie(cost_dist, labels = cost_dist.index, autopct = "%1.2f%%")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
payment_counts = df['Payment Method'].value_counts()
plt.pie(payment_counts, labels=payment_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title("Payment Method Proportion")
plt.show()
```

```python
# Assuming 'Order Date and Time' is already in datetime format
df['Order Date'] = df['Order Date and Time'].dt.date
daily_order_value = df.groupby('Order Date')['Order Value'].sum()

plt.figure(figsize=(12, 6))
plt.plot(daily_order_value.index, daily_order_value.values, marker='o', color='purple')
plt.title("Total Order Value Over Time")
plt.xlabel("Order Date")
plt.ylabel("Total Order Value")
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
delivery_fee_by_payment = df.groupby(['Payment Method'])['Delivery Fee'].sum()

plt.figure(figsize=(10, 6))
delivery_fee_by_payment.plot(kind='bar', stacked=True, color=["#FF6F61", "#5D9CEC", "#6B8E23"])
plt.title("Total Delivery Fees by Payment Method")
plt.xlabel("Payment Method")
plt.ylabel("Total Delivery Fee")
plt.xticks(rotation=45)
plt.show()
```

```python
abc = df[["Commission Fee", "Costs","Profit"]].sum()
abc
```

```python
plt.figure(figsize = (3,3))
plt.bar(abc.index, abc)
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
payment_counts = df['Payment Method'].value_counts()
payment_counts.plot(kind='bar', color=['#5D9CEC', '#FF6F61', '#FFE156'])
plt.title("Payment Method Breakdown")
plt.xlabel("Payment Method")
plt.ylabel("Count of Orders")
plt.xticks(rotation=45)
plt.show()
```

```python
avg_order_value_by_restaurant = df.groupby('Restaurant ID')['Order Value'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
avg_order_value_by_restaurant.plot(kind='bar', color="teal")
plt.title("Top 10 Restaurants by Average Order Value")
plt.xlabel("Restaurant ID")
plt.ylabel("Average Order Value")
plt.xticks(rotation=45)
plt.show()
```
